```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
sales_by_month = pd.read_csv('sales_by_month.csv')
sales_by_product_category = pd.read_csv('sales_by_product_category.csv')
top_selling_products = pd.read_csv('top_selling_products.csv')
customer_demographics = pd.read_csv('customer_demographics.csv')
sales_channels = pd.read_csv('sales_channels.csv')

# Sales by Month
plt.figure(figsize=(10, 6))
plt.plot(sales_by_month['Month'], sales_by_month['Sales Revenue'], marker='o')
plt.title('Sales Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Sales Revenue ($)')
plt.grid(True)
plt.savefig('sales_by_month.png')
plt.show()



print("\nSales by Month:\n")
print(sales_by_month)
```

```python
# Sales by Product Category
plt.figure(figsize=(10, 6))
plt.pie(sales_by_product_category['Percentage of Total Sales'], labels=sales_by_product_category['Category'], autopct='%1.1f%%')
plt.title('Sales by Product Category')
plt.savefig('sales_by_product_category.png')
plt.show()



print("\nSales by Product Category:\n")
print(sales_by_product_category)
```

```python
# Top Selling Products
plt.figure(figsize=(10, 6))
plt.bar(top_selling_products['Product'], top_selling_products['Units Sold'], color='skyblue')
plt.title('Top Selling Products by Units Sold')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.savefig('top_selling_products.png')
plt.show()


print("\nTop Selling Products:\n")
print(top_selling_products)
```

```python
# Customer Demographics
plt.figure(figsize=(10, 6))
age_groups = customer_demographics['Age Group']
male = customer_demographics['Male']
female = customer_demographics['Female']

bar_width = 0.35
index = range(len(age_groups))

plt.bar(index, male, bar_width, label='Male')
plt.bar(index, female, bar_width, bottom=male, label='Female')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.title('Customer Demographics by Age and Gender')
plt.xticks(index, age_groups)
plt.legend()
plt.savefig('customer_demographics.png')
plt.show()



print("\nCustomer Demographics:\n")
print(customer_demographics)
```

```python

# Sales Channels
plt.figure(figsize=(10, 6))
plt.pie(sales_channels['Percentage of Total Sales'], labels=sales_channels['Channel'], autopct='%1.1f%%')
plt.title('Sales by Channel')
plt.savefig('sales_channels.png')
plt.show()



print("\nSales Channels:\n")
print(sales_channels)
```
